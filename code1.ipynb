```python
import matplotlib.pyplot as plt
import numpy as np

# Define the coordinates for the base station, access point, and IRS device
base_station = np.array([0, 0])         # Base station at (0, 0)
access_point = np.array([50, 0])        # Access point 50 meters to the right on the same line
irs_device = np.array([25, 2])          # IRS device 2 meters above and at the midpoint between the base station and access point

# Calculate the distances of the triangle
distance_base_to_irs = np.linalg.norm(irs_device - base_station)
distance_access_to_irs = np.linalg.norm(irs_device - access_point)
distance_base_to_access = np.linalg.norm(access_point - base_station)

# Plot the points
plt.figure(figsize=(8, 6))

# Plot the base station, access point, and IRS device with different markers
plt.plot(*base_station, 'ro', label='Base Station (●)')
plt.plot(*access_point, 'bo', label='Access Point (■)')
plt.plot(*irs_device, 'g^', label='IRS Device (▲)')

# Draw the horizontal lines
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5) # Line for base station and access point
plt.axhline(y=2, color='black', linestyle='--', linewidth=0.5) # Line for IRS device (2 meters above)

# Join the points to form a triangle
plt.plot([base_station[0], irs_device[0]], [base_station[1], irs_device[1]], 'k-', label=f'Distance Base to IRS: {distance_base_to_irs:.2f} m')
plt.plot([access_point[0], irs_device[0]], [access_point[1], irs_device[1]], 'k-', label=f'Distance Access to IRS: {distance_access_to_irs:.2f} m')
plt.plot([base_station[0], access_point[0]], [base_station[1], access_point[1]], 'k-', label=f'Distance Base to Access: {distance_base_to_access:.2f} m')

# Set axis limits
plt.xlim(-10, 60)
plt.ylim(-5, 10)

# Add labels and title
plt.xlabel('X-axis (meters)')
plt.ylabel('Y-axis (meters)')
plt.title('Base Station, Access Point, and IRS Device Configuration')
plt.grid(True)

# Add legend
plt.legend()

# Show plot
plt.show()
```
