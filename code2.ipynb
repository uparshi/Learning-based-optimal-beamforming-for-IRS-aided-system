```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
fc = 2  # Frequency in GHz
distance_values = np.linspace(10, 50, 5)  # Distance values in meters (you can customize this range)

# LOS calculation function
def calculate_los(fc, distance):
    return -28 - 20 * np.log10(fc) - 22 * np.log10(distance)

# NLOS calculation function
def calculate_nlos(fc, distance):
    return -22.7 - 26 * np.log10(fc) - 36.7 * np.log10(distance)

# Calculate LOS and NLOS for each distance
los_values = calculate_los(fc, distance_values)
nlos_values = calculate_nlos(fc, distance_values)

# Plotting the results
plt.figure(figsize=(8, 6))
plt.plot(distance_values, los_values, marker='o', linestyle='-', color='b', label='LOS')
plt.plot(distance_values, nlos_values, marker='x', linestyle='--', color='r', label='NLOS')
plt.title('Direct LOS and NLOS vs Distance')
plt.xlabel('Distance (m)')
plt.ylabel('Signal Strength (dB)')
plt.grid(True)
plt.legend()
plt.show()

# Show the LOS and NLOS values for the distances
for dist, los, nlos in zip(distance_values, los_values, nlos_values):
    print(f"Distance: {dist} meters, Direct LOS: {los:.2f} dB, NLOS: {nlos:.2f} dB")
```
